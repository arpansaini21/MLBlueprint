### 1. Vector Creation  Numpy Array 
<a href="#0">Click here to Go to Table of content</a>

```python
def print_numpy_array(headline,arr):
    print(headline)
    print(f"array :   a = {arr}, a shape = {arr.shape}, a data type = {arr.dtype}",'\n')
```

```python
import numpy as np

a = np.zeros(4) 
print_numpy_array("# passing singluar dimention",a)


a = np.zeros((4,)) 
print_numpy_array('# passing tuple as dimention (n,) -> 1-D , (n,n) -> 2 -D etc', a)



a = np.arange(8.0).reshape((2,4))
a = np.zeros_like(a)
print_numpy_array('#create zero array, copy shape and data type of passed array', a)


a = np.random.random_sample((4,)) 
print_numpy_array('# all random samples has values less than one',a)


a = np.arange(4)
print_numpy_array('# range you can pass integer or float , Not taking dimention as input, Create 1-D array', a) 


a = np.random.rand(4) 
print_numpy_array('# passing the value not as dimention ( always create 1-D Array )', a)


a = np.array([1,4,7,9]) 
print_numpy_array('# passing data manually to create an array', a)

a = np.array([[1,4,7,9],[4,5,6,7]]) 
print_numpy_array('',a)

a = np.array([1.0,4,7,9]) 
print_numpy_array('# passing data manualy but list contdain different datatypes', a)
```

### 2. Indexing and Slicing
- Indexing : getting single element from an array
- Slicing : getting subset of element from an array `(start-include:stop-depend on step:step-1 means consequitive)`

<a href="#0">Click here to Go to Table of content</a>

```python
#vector indexing operations on 1-D vectors
a = np.arange(10)
print(a)
print(a[0],a[-10], a[9], a[-1])

try:
     c = a[10]
except Exception as e:
    print("The error message you will see is")
    print(e)

    
    
#access 5 consecutive elements (start:stop:step) ( start and end are included )
c = a[2:7:1];  
print("a[2:7:1] = ", c)

# access 3 elements separated by two 
c = a[2:7:2];     
print("a[2:7:2] = ", c)

# access all elements index 3 and above ( start index include )
c = a[3:];        
print("a[3:]    = ", c)

# access all elements below index 3 ( end endex is not included )
c = a[:3];        
print("a[:3]    = ", c)

# access all elements
c = a[:];         
print("a[:]     = ", c)
```

### 5. Vector Vector dot product
<a href="#0">Click here to Go to Table of content</a>

```python
import numpy as np
import time
def my_dot(a, b): 
    """
   Compute the dot product of two vectors
 
    Args:
      a (ndarray (n,)):  input vector 
      b (ndarray (n,)):  input vector with same dimension as a
    
    Returns:
      x (scalar): 
    """
    x=0
    for i in range(a.shape[0]):
        x = x + a[i] * b[i]
    return x
```

```python
# test 1-D
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
print(f"my_dot(a, b) = {my_dot(a, b)}")
```

```python
# test 1-D
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
c = np.dot(a, b)
print(f"NumPy 1-D np.dot(a, b) = {c}, np.dot(a, b).shape = {c.shape} ") 
c = np.dot(b, a)
print(f"NumPy 1-D np.dot(b, a) = {c}, np.dot(a, b).shape = {c.shape} ")
```

#### Vector speed test

```python
np.random.seed(1)
a = np.random.rand(10000000)  # very large arrays
b = np.random.rand(10000000)

tic = time.time()  # capture start time
c = np.dot(a, b)
toc = time.time()  # capture end time

print(f"np.dot(a, b) =  {c:.4f}")
print(f"Vectorized version duration: {1000*(toc-tic):.4f} ms ")

tic = time.time()  # capture start time
c = my_dot(a,b)
toc = time.time()  # capture end time

print(f"my_dot(a, b) =  {c:.4f}")
print(f"loop version duration: {1000*(toc-tic):.4f} ms ")

del(a);del(b)  #remove these big arrays from memory
```

### 7. Mash grid
<a href="#0">Click here to Go to Table of content</a>

- create all possible pairs of two 1-D Array
- It turns 1D vectors into 2D coordinate matrices.
- Used in ML to plot the cost function for every pair of w,b

```python
w_vals = np.array([1,2,3])
b_vals = np.array([5,6])

# with these two array, you can create all the pairs of x and y
W, B = np.meshgrid(w_vals, b_vals)
print(W,B)
```
